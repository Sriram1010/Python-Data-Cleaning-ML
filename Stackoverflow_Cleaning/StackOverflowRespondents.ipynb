```python
import pandas as pd
import numpy as np
```

```python
sodf16 = pd.read_csv('survey_results_public2016.csv', usecols = ['Respondent', 'country', 'age_midpoint', 'salary_midpoint'])
sodf16.rename(columns = {'country' : 'Country', 'age_midpoint' : 'Age', 'salary_midpoint' : 'Salary'}, inplace = True)
sodf16['Year'] = 2016
display(sodf16)
```

```python
sodf17 = pd.read_csv('survey_results_public2017.csv', usecols = ['Respondent', 'Country', 'Salary'])
sodf17['Year'] = 2017
display (sodf17)
```

```python
sodf18 = pd.read_csv('survey_results_public2018.csv', usecols = ['Respondent', 'Country', 'Age','ConvertedSalary'])
sodf18['Year'] = 2018
sodf18.rename(columns = {'ConvertedSalary' : 'Salary'}, inplace = True)
sodf18.replace(['Under 18 years old', '18 - 24 years old', '25 - 34 years old', '35 - 44 years old',
                      '45 - 54 years old', '55 - 64 years old', '65 years or older'],
                    [15, 21, 29.5, 39.5, 49.5 , 59.5, 70], inplace = True)
display(sodf18)
```

```python
sodf19 = pd.read_csv('survey_results_public2019.csv', usecols = ['Respondent', 'Country', 'ConvertedComp', 'Age'])
sodf19.rename(columns = {'ConvertedComp' : 'Salary'}, inplace = True)
sodf19['Year'] = 2019
display(sodf19)
```

```python
sodf20 = pd.read_csv('survey_results_public2020.csv', usecols = ['Respondent', 'Country', 'Age', 'ConvertedComp'])
sodf20.rename(columns = {'ConvertedComp' : 'Salary'}, inplace = True)
sodf20['Year'] = 2020
display(sodf20)
```

```python
sodf = sodf16.append([sodf17, sodf18, sodf19, sodf20])
sodf['Salary'] = sodf['Salary'].round(0)
display(sodf)
```

```python
sodfUnique = sodf[['Country', 'Year']].drop_duplicates()
sodfGroupBy = sodfUnique.groupby('Country')['Year'].apply(list)
sodfGroupBy = sodfGroupBy.reset_index()
sodfGroupBy.sort_values(by = ['Country'])
display(sodfGroupBy)
```

```python
sodfGroupBy.to_csv('StackOverflowCountries.csv', index = False)
```
