```python
import numpy as np
import pandas as pd
```

```python
sodf16 = pd.read_csv("survey_results_public2016.csv", usecols = ['Respondent', 'gender'])
sodf16.rename(columns = {'gender' : 'Gender'}, inplace = True)
sodf16.dropna(inplace = True)
sodf16['Year'] = 2016
display(sodf16)
```

```python
sodf17 = pd.read_csv("survey_results_public2017.csv", usecols = ['Respondent', 'Gender'])
sodf17.dropna(inplace = True)
sodf17['Year'] = 2017
display(sodf17)
```

```python
sodf17['Gender'] = sodf17['Gender'].str.split('; ')
sodf17Explode = sodf17.explode('Gender')
display (sodf17Explode)
```
