```python
import numpy as np
import pandas as pd
```

```python
techdf = pd.read_excel("DimTechnology_Updated_V2.xlsx")
techdf2016 = techdf.loc[techdf['Year'] == 2016]
display(techdf2016)
```

```python
sodf16 = pd.read_csv("survey_results_public2016.csv", usecols = ['Respondent', 'tech_do', 'dev_environment', 'desktop_os'])
sodf16.rename(columns = {"tech_do" : "Tech", "dev_environment" : "DevEnv", "desktop_os": "OpSys"}, inplace = True)
sodf16['Year'] = 2016
display(sodf16)
```

```python
sodf16.dropna(subset=['Tech', 'DevEnv', 'OpSys'], how = 'all', inplace = True)
display(sodf16)
```

```python
sodf16Melt = sodf16.melt(id_vars = ['Respondent', 'Year'], 
                         value_vars = ['Tech', 'DevEnv', 'OpSys'], 
                         var_name = 'Category', value_name = 'Technology')
sodf16Melt.dropna(inplace = True)
display(sodf16Melt)
```

```python
sodf16Melt['Technology'] = sodf16Melt['Technology'].str.split('; ') 
display(sodf16Melt)
```

```python
sodf16Explode = sodf16Melt.explode('Technology', ignore_index = True)
display(sodf16Explode)
```

```python
sodf16Explode['Technology'].replace({'Arduino / Raspberry Pi': 'Arduino/Raspberry Pi',
                                    'SQL Server' : 'Microsoft SQL Server',
                                    'Visual Basic': 'Visual Basic 6'}, inplace = True)
sodf16Merge = sodf16Explode.merge(right = techdf2016[['Category', 'Technology']], how = 'inner', on = 'Technology')
display(sodf16Merge)
```

```python
sodf16Merge.drop(columns = ['Category_x'], inplace = True)
sodf16Merge.rename(columns = {"Category_y": "Category"}, inplace = True)
display(sodf16Merge)
```

```python
sodf16Merge.drop_duplicates(inplace = True)
display (sodf16Merge)
```

```python
sodf17 = pd.read_csv("survey_results_public2017.csv",
                    usecols = ['Respondent','HaveWorkedLanguage', 'HaveWorkedDatabase', 'HaveWorkedFramework','HaveWorkedPlatform', 'IDE'])
sodf17.rename(columns = {"HaveWorkedLanguage": "Language", "HaveWorkedDatabase": "Database",
                         "HaveWorkedFramework" : "Framework", "HaveWorkedPlatform" : "Platform",
                        "IDE": "DevEnv"}, inplace = True)
sodf17.dropna(subset =['Language', 'Framework', 'Database', 'Platform', 'DevEnv'], how = 'all', inplace = True)
sodf17['Year'] = 2017
display(sodf17)
```

```python
sodf18 = pd.read_csv("survey_results_public2018.csv", usecols = ['Respondent','LanguageWorkedWith','DatabaseWorkedWith',
                 'FrameworkWorkedWith','PlatformWorkedWith', 'IDE', 'OperatingSystem'])
sodf18.rename(columns = {'LanguageWorkedWith' : 'Language',
                         'DatabaseWorkedWith' : 'Database',
                         'FrameworkWorkedWith' : 'Framework',
                         'PlatformWorkedWith' : 'Platform',
                         'IDE' : 'DevEnv',
                         'OperatingSystem' : 'OpSys'}, inplace = True)
sodf18.dropna(subset = ['Language', 'Database', 'Platform', 'Framework', 'DevEnv', 'OpSys'], how = 'all', inplace = True)
sodf18['Year'] = 2018
display(sodf18)
```

```python
sodf19 = pd.read_csv("survey_results_public2019.csv", usecols = ['Respondent', 'LanguageWorkedWith', 'DatabaseWorkedWith',
                 'WebFrameWorkedWith','PlatformWorkedWith', 'MiscTechWorkedWith', 'DevEnviron', 'OpSys'])
sodf19.rename(columns = {'LanguageWorkedWith' : 'Language', 'DatabaseWorkedWith' : 'Database', 'WebFrameWorkedWith' : 'Webframe',
                        'PlatformWorkedWith' : 'Platform', 'MiscTechWorkedWith' : 'MiscTech', 'DevEnviron' : 'DevEnv'}, inplace = True)
sodf19['Year'] = 2019
sodf19.dropna(subset = ['Language', 'Database', 'Platform', 'Webframe','MiscTech', 'DevEnv', 'OpSys'], how = 'all', inplace = True)
display(sodf19)
```

```python
sodf20 = pd.read_csv("survey_results_public2020.csv",
                     usecols = ['Respondent','LanguageWorkedWith',
                                'DatabaseWorkedWith','WebframeWorkedWith',
                                'PlatformWorkedWith','MiscTechWorkedWith',
                                'NEWCollabToolsWorkedWith', 'OpSys'])
sodf20.rename(columns = {'LanguageWorkedWith' : 'Language', 'DatabaseWorkedWith' : 'Database', 'WebframeWorkedWith' : 'Webframe',
                        'PlatformWorkedWith' : 'Platform', 'MiscTechWorkedWith' : 'MiscTech', 'NEWCollabToolsWorkedWith' : 'CollabTools'},
             inplace = True)
sodf20['Year'] = 2020
sodf20.dropna(subset = ['Database', 'Language', 'MiscTech', 'CollabTools', 'OpSys', 'Platform', 'Webframe'], how = 'all', inplace = True)
display(sodf20)
```

```python
sodf = sodf17.append([sodf18, sodf19, sodf20])
display(sodf)
```

```python
sodfMelt = sodf.melt(id_vars = ['Respondent', 'Year'], var_name = 'Category', value_name = 'Technology')
sodfMelt.dropna(inplace = True)
sodfMelt['Technology'] = sodfMelt['Technology'].str.split(';')
display(sodfMelt)
```

```python
sodfExplode = sodfMelt.explode('Technology', ignore_index = True)
display(sodfExplode)
```

```python
sodfExplode['Technology'] = sodfExplode['Technology'].str.strip()
```

```python
sodfFinal = sodfExplode.append(sodf16Merge)
sodfFinal['Technology'].replace('Other(s):', 'Other', inplace = True)
sodfFinal= sodfFinal[sodfFinal['Technology'] != 'Other']
sodfFinal['Technology'].replace('Amazon DynamoDB', 'DynamoDB', inplace = True)
sodfFinal['Technology'].replace('Angular.js', 'AngularJS', inplace = True)
sodfFinal['Technology'].replace('Angular/Angular.js', 'Angular/AngularJS', inplace = True)
sodfFinal['Technology'].replace('AWS', 'Amazon Web Services (AWS)', inplace = True)
sodfFinal['Technology'].replace('Azure', 'Microsoft Azure', inplace = True)
sodfFinal['Technology'].replace('Google Suite (Docs, Meet, etc)', 'Google Suite (Docs, Meet, etc.)', inplace = True)
sodfFinal['Technology'].replace('IBM DB2', 'IBM Db2', inplace = True)
sodfFinal['Technology'].replace('Lighttable', 'Light Table', inplace = True)
sodfFinal['Technology'].replace('Linux Desktop', 'Linux', inplace = True)
sodfFinal['Technology'].replace('Mac OS', 'MacOS', inplace = True)
sodfFinal['Technology'].replace('Microsoft Azure (Tables, CosmosDB, SQL, etc)', 'Microsoft Azure (Tables, CosmosDB, SQL, etc.)', inplace = True)
sodfFinal['Technology'].replace(['React','React.js'], 'ReactJS', inplace = True)

sodfFinal['Technology'].replace('SQL Server', 'Microsoft SQL Server', inplace = True)
sodfFinal['Technology'].replace(['Windows Desktop', 'Windows Desktop or Server'], 'Windows', inplace = True)
sodfFinal.loc[sodfFinal['Technology'] == 'Node.js', 'Category'] = 'MiscTech'
sodfFinal.loc[sodfFinal['Technology'] == 'Apache Spark', 'Category'] = 'Framework'
sodfFinal.sort_values(by = ['Year', 'Respondent'], inplace = True)
display(sodfFinal)
```

```python
sodfFinal.to_csv('StackOverflowRespondentTechnologiesv2.csv', index = False)
```

```python
sodfTechnologies = sodfFinal[['Technology', 'Category']].copy()
sodfTechnologies.drop_duplicates(inplace = True)
sodfTechnologies.to_csv('SOTechnologiesDuplicates.csv', index = False)
```
